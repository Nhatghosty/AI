# Feature Engineering
Create new features such as 'hour', 'day_of_week', 'is_weekend', and calculate the distance between charging sessions.

```python
# Extract hour, day_of_week, and is_weekend from 'Charging Start Time'
df['Charging Start Time'] = pd.to_datetime(df['Charging Start Time'], errors='coerce')
df['hour'] = df['Charging Start Time'].dt.hour
df['day_of_week'] = df['Charging Start Time'].dt.dayofweek
df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)

# Calculate the distance between charging sessions
if 'Distance Driven' in df.columns and 'Vehicle ID' in df.columns:
    df = df.sort_values(by=['Vehicle ID', 'Charging Start Time'])
    df['Distance Between Charges'] = df.groupby('Vehicle ID')['Distance Driven'].diff().fillna(0)
else:
    df['Distance Between Charges'] = None

# Inspect the new features
print("\nNew features added to the dataset:")
print(df[['hour', 'day_of_week', 'is_weekend', 'Distance Between Charges']].head())
```

# Train Logistic Regression Model
Train the Logistic Regression model using cross-validation and optimize hyperparameters using GridSearchCV.

```python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, train_test_split
from category_encoders import TargetEncoder

# Load the dataset
file_path = 'ev_charging_patterns.csv'  # Đảm bảo đường dẫn chính xác
df = pd.read_csv(file_path)

# Separate features and target variable
X = df.drop(columns=["User Type"])
y = df["User Type"]

# Encode target variable for Logistic Regression
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Apply Target Encoding for Logistic Regression
categorical_columns = X.select_dtypes(include=['object']).columns
target_encoder = TargetEncoder(cols=categorical_columns)
X_encoded_logistic = target_encoder.fit_transform(X, y_encoded)

# Chuẩn hóa dữ liệu
standard_scaler = StandardScaler()
X_standard_scaled = standard_scaler.fit_transform(X_encoded_logistic)

# Chia tập dữ liệu
X_train, X_test, y_train, y_test = train_test_split(
    X_standard_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print("Train-test split completed successfully.")

# Update the pipeline to include an imputer
logistic_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),  # Handle missing values
    ('scaler', StandardScaler()),
    ('model', LogisticRegression(random_state=42, max_iter=1000))
])

# Define the parameter grid for GridSearchCV
param_grid = {
    'model__C': [0.1, 1, 10],  # Example hyperparameters for Logistic Regression
    'model__solver': ['lbfgs', 'liblinear']
}

# Proceed with GridSearchCV
logistic_grid_search = GridSearchCV(
    estimator=logistic_pipeline,
    param_grid=param_grid,
    cv=5,
    scoring='f1_weighted',
    n_jobs=-1
)

logistic_grid_search.fit(X_train, y_train)
print("GridSearchCV for Logistic Regression completed successfully.")
```

# Train CatBoost Model
Train the CatBoost model using cross-validation and optimize hyperparameters using Optuna.

```python
import optuna
from sklearn.model_selection import cross_val_score
from catboost import CatBoostClassifier

# Define objective function for CatBoost optimization
def objective_catboost(trial):
    iterations = trial.suggest_int('iterations', 100, 300)  # Giảm phạm vi
    depth = trial.suggest_int('depth', 4, 8)  # Giảm độ sâu
    learning_rate = trial.suggest_float('learning_rate', 0.01, 0.1, log=True)  # Thu hẹp phạm vi
    l2_leaf_reg = trial.suggest_float('l2_leaf_reg', 1, 5, log=True)  # Thu hẹp phạm vi

    model = CatBoostClassifier(
        iterations=iterations,
        depth=depth,
        learning_rate=learning_rate,
        l2_leaf_reg=l2_leaf_reg,
        random_state=42,
        verbose=0,
        early_stopping_rounds=50  # Dừng sớm
    )

    scores = cross_val_score(model, X_train, y_train, cv=5, scoring='f1_weighted')
    return scores.mean()

# Optimize CatBoost hyperparameters using Optuna
catboost_study = optuna.create_study(direction='maximize')
catboost_study.optimize(objective_catboost, n_trials=10, timeout=3600)  # Giảm số thử nghiệm và giới hạn thời gian
```

# Model Evaluation
Evaluate both models using metrics like accuracy, precision, recall, F1-score, and confusion matrix.

```python
# Import f1_score nếu chưa được nhập
from sklearn.metrics import f1_score

# Evaluate Logistic Regression on the test set
logistic_test_predictions = logistic_grid_search.best_estimator_.predict(X_test)
logistic_test_f1 = f1_score(y_test, logistic_test_predictions, average='weighted')

# Evaluate CatBoost on the test set
best_catboost_model = CatBoostClassifier(**catboost_study.best_params, random_state=42, verbose=0)
best_catboost_model.fit(X_train, y_train)
catboost_test_predictions = best_catboost_model.predict(X_test)
catboost_test_f1 = f1_score(y_test, catboost_test_predictions, average='weighted')

print(f"Logistic Regression Test F1-score: {logistic_test_f1}")
print(f"CatBoost Test F1-score: {catboost_test_f1}")
```

# Compare Models and Select the Best
Compare the performance of Logistic Regression and CatBoost models and select the best one based on F1-score.

```python
# Compare F1-scores and select the best model
if logistic_test_f1 > catboost_test_f1:
    print("Logistic Regression is the better model based on F1-score.")
    best_model = logistic_grid_search.best_estimator_
else:
    print("CatBoost is the better model based on F1-score.")
    best_model = best_catboost_model
```
